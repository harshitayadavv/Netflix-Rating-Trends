```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('data/Netflix TV Shows and Movies.csv')
df
```

```python
# 1. Print the first 10 rows from the dataset
# Display the first 10 rows to understand the dataset structure
df.head(10)
```

```python
# 2. Check the number of rows and columns in the dataset
# Check the shape of the dataset (number of rows and columns)
df.shape
```

```python
# 3. Check the type of each column in the dataset
# Display data types for each column
df.dtypes
```

```python
# 4. Number of unique values in each column
# Count the number of unique values for each column
df.nunique()
```

```python
# 5. Remove instances with missing values in critical columns
# Removing rows where title or imdb_score is missing (they are critical for analysis)
df = df.dropna(subset=['title', 'imdb_score'])
df
```

```python
# 6. Descriptive Statistics of the Dataset (count, mean, std, min, 25%, 50%, 75%, max)
# Get summary statistics for numeric columns
df.describe()
```

```python
# 7. **Check for duplicates in the dataset**
# Use the duplicated() method to check for duplicate rows in the dataset
duplicates = df.duplicated()  # Returns a boolean Series, True if the row is a duplicate, False otherwise

# To see the number of duplicate rows
num_duplicates = duplicates.sum()  # Sums up the number of True values (duplicate rows)

# To view the actual duplicate rows
duplicate_rows = df[df.duplicated()]  # Returns all rows that are duplicates

# Output the results
print(f"Number of duplicate rows: {num_duplicates}")
print("Duplicate Rows:")
print(duplicate_rows)
```

```python
# --- Analysis & Filtering ---

# 8. **Identify unique types in the dataset (Movie or Show)**
df['type'].value_counts()  # Shows how many movies and shows are in the dataset
```

```python
# 9. **Identify unique age certifications and print their counts**
df['age_certification'].value_counts()  # Counts the number of unique age certifications
```

```python
# 10. **Print only movies which IMDb Score is greater than 7**
df_movies_above_7 = df[df['imdb_score'] > 7]  # Filters movies with IMDb score greater than 7
df_movies_above_7
```

```python
# 11. **Print only movies which IMDb Score is less than 4**
df_movies_below_4 = df[df['imdb_score'] < 4]  # Filters movies with IMDb score less than 4
df_movies_below_4
```

```python
# 12. **Top 10 movies with IMDb Score greater than 7**
top_10_movies_above_7 = df_movies_above_7.nlargest(10, 'imdb_score')  # Shows top 10 movies with IMDb score > 7
top_10_movies_above_7[['title', 'imdb_score']]  # Display only the movie titles and IMDb scores
```

```python
# 13. **Calculate the average runtime of movies in the dataset**
average_runtime = df['runtime'].mean()  # Calculates the average runtime of all movies
average_runtime
```

```python
# 14. **Print only movies which IMDb Score is greater than 7 and runtime is less than 60 minutes**
df_movies_above_7_runtime_below_60 = df[(df['imdb_score'] > 7) & (df['runtime'] < 60)]  # Filters based on IMDb score and runtime
df_movies_above_7_runtime_below_60[['title', 'imdb_score', 'runtime']]  # Display the filtered movies
```

```python
# 15. **Print only the name of movies and their type where IMDb Score is greater than 7 and runtime is less than 60 minutes**
df_movies_above_7_runtime_below_60[['title', 'type']]  # Displays only movie name and type
```

```python
# 16. **List IMDb Movie ratings that are less than 5 and their respective types**
df_movies_below_5 = df[df['imdb_score'] < 5]  # Filters movies with IMDb score below 5
df_movies_below_5[['title', 'imdb_score', 'type']]  # Display the movie names, ratings, and types
```

```python
# 17. **How many movies premiered between 2010 and 2015?**
df_movies_2010_2015 = df[(df['release_year'] >= 2010) & (df['release_year'] <= 2015)]  # Filters based on release year
df_movies_2010_2015.shape[0]  # Returns the number of movies in this range
```

```python
# 18. Sort the movies based on IMDb Score in descending order
sorted_movies = df.sort_values(by='imdb_score', ascending=False)
sorted_movies
```

```python
# --- Data Visualisation ---

# 1. **Movies per year (Bar Chart)**
import matplotlib.pyplot as plt
df['release_year'].value_counts().sort_index().plot(kind='bar', figsize=(10,6))
plt.title('Movies Per Year')
plt.xlabel('Year')
plt.ylabel('Count of Movies')
plt.show()
```

```python
# 2. **IMDb Score Distribution (Histogram or Boxplot)**
df['imdb_score'].plot(kind='hist', bins=20, figsize=(10,6), color='skyblue')
plt.title('IMDb Score Distribution')
plt.xlabel('IMDb Score')
plt.ylabel('Count')
plt.show()
```

```python
# 3. **Average IMDb Score per Movie Type (Bar Chart)**
df.groupby('type')['imdb_score'].mean().plot(kind='bar', figsize=(10,6), color='salmon')
plt.title('Average IMDb Score per Movie Type')
plt.xlabel('Type')
plt.ylabel('Average IMDb Score')
plt.show()
```

```python
# 4. **Movies per Age Certification (Bar Chart)**
df['age_certification'].value_counts().plot(kind='bar', figsize=(10,6), color='lightgreen')
plt.title('Movies per Age Certification')
plt.xlabel('Age Certification')
plt.ylabel('Count of Movies')
plt.show()
```

```python
# 5. **Movies by Runtime (Histogram)**
df['runtime'].plot(kind='hist', bins=20, figsize=(10,6), color='purple')
plt.title('Movies by Runtime')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Count of Movies')
plt.show()
```
